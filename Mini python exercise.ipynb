```python
# -------------------------------
# Mini Python Practice Project
# -------------------------------

# 1️⃣ Password Verification System
def password_verification():
    password = 1234
    enter_password = int(input("Enter password: "))
    while enter_password != password:
        print("Incorrect password, try again!")
        enter_password = int(input("Enter password: "))
    print("Login successful!\n")

# 2️⃣ Multiplication Tables Using Nested Loops
def multiplication_tables():
    print("Multiplication Tables (1-3):")
    for i in range(1, 4):        # tables for 1 to 3
        for j in range(1, 11):   # multiply by 1 to 10
            print(f"{i} * {j} = {i*j}")
        print()  # empty line between tables

# 3️⃣ Grade Calculator Function
def grade_calculator(marks):
    if marks >= 90:
        print("Grade A")
    elif marks >= 80:
        print("Grade B")
    elif marks >= 70:
        print("Grade C")
    else:
        print("Failed")

# 4️⃣ Bank Account Simulation with Global Variables
balance = 1000  # global variable

def deposit(amount):
    global balance
    balance += amount
    print(f"Deposited: {amount}")
    print(f"New Balance: {balance}")

def withdraw(amount):
    global balance
    if amount <= balance:
        balance -= amount
        print(f"Withdrew: {amount}")
        print(f"New Balance: {balance}")
    else:
        print("Insufficient balance!\n")

# -------------------------------
# Running the Mini Project
# -------------------------------

# Password system
password_verification()

# Multiplication tables
multiplication_tables()

# Grade calculator examples
print("Grade Examples:")
grade_calculator(100)
grade_calculator(75)
grade_calculator(50)
print()

# Bank account transactions
print("Bank Account Simulation:")
withdraw(500)
deposit(1500)
withdraw(2000)  # attempt to withdraw more than balance
```
