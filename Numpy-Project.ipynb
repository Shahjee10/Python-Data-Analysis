```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#step 1 generate synthetic sales data
np.random.seed(42) # for reproducibility
months = np.array(["Jan","Feb","Mar","Apr","May","Jun",
                   "Jul","Aug","Sept","Oct","Nov","Dec"])

sales = np.random.randint(200, 1000, size = 12) #random sales between 200 and 1000
```

```python
# ---------------- STEP 2: Simple analysis ----------------

# Total sales for the year
total_sales = np.sum(sales)

# Average sales per month
avg_sales = np.mean(sales)

# Find index of best and worst months
best_month = months[np.argmax(sales)]   # month with highest sales
worst_month = months[np.argmin(sales)]  # month with lowest sales

# Print results
print("Total Sale:", total_sales)
print("Average Sale:", avg_sales)
print("Best Month:", best_month, "with sales", np.max(sales))
print("Worst Month:", worst_month, "with sales", np.min(sales))
```

```python
# ---------------- STEP 3: Visualize with charts ----------------

# Create a figure with size
plt.figure(figsize=(14,5))

# ---- Bar Chart (left side) ----
plt.subplot(1,2,1)  # 1 row, 2 columns, first plot
plt.bar(months, sales, color="skyblue")  # draw bar chart
plt.title("Monthly Sales (Bar Chart)")
plt.xlabel("Months")
plt.ylabel("Sales")

# ---- Line Chart (right side) ----
plt.subplot(1,2,2)  # 1 row, 2 columns, second plot
plt.plot(months, sales, marker="o", color="skyblue", linestyle="--")  # draw line chart
plt.title("Monthly Sales (Line Chart)")
plt.xlabel("Months")
plt.ylabel("Sales")

# Adjust layout so labels don't overlap
plt.tight_layout()

# Show the figure
plt.show()
```
